## Practicing Tokenization using Regular Expressions

```python
# Open "The Verdict"
with open("./the-verdict.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
print(f"Total No. Words: {len(raw_text)}")
```

```python
# Tokenize the total no. words into individual words + spaecial characters
import re

print("Sample with spaces: \n{s}\n\n".format(s=re.split(r'(\s)', raw_text)))
print("Sample with spaces and punctuation:\n{s}\n\n".format(s=re.split(r'([,.]|\s)', raw_text)))
print("Sample without white spaces and puntuation:\n{s}\n\n".format(s=[item for item in re.split(r'([,.]|\s)', raw_text) if item.strip()]))
print("Sample without white spaces and additional punctuation:\n\n{s}\n".format(s=[item for item in re.split(r'([,.;:?!_"()]|--|\s)', raw_text) if item.strip()]))
```

```python
# Converting tokens to tokoen IDs
tokenized_text = re.split(r'([.,?!:;_()"\']|--|\s)', raw_text)
tokenized_text = [item.strip() for item in tokenized_text if item.strip()]
```

```python
# Sort unique occurances of eaach token in ascednding order
sorted_text = sorted(set(tokenized_text))
print(f"Vocabulary Size: {len(sorted_text)}")
```

```python
# Build the vocabulary --> Dictionary of unique token and token ID's
vocab = {token:integer for integer, token in enumerate(sorted_text)}
```

```python
# Vire the vocabulary
for index, item in enumerate(vocab.items()):
    print(item)
    if index >= 50:
        break
```

## Tokenizer Class to Instantiate new Tokenizer Objects

The tokenizer implementation has two specific methods, `encoder`, and `decoder`. The `encoder` method takes in sample text, splits it into individual tokens, anmd then converts the the individual tokens into token IDs, via the vocabulary. The `decode` method takes the token IDs and converts them back into individual text tokens . It them concatenates the text tokens into natual text, taking spaces and punctuation into account.

```python
class SimpleTokenizerV1:
    def __init__(self, vocab):
        # Initialize the class attribute for the vocabulary
        self.str_to_int = vocab

        # Initialize the inverse vocabulary to map toekn IDs back to the original text tokens
        self.int_to_str = {integer:string for string, integer in vocab.items()}
    
    # Create the `encode` method to process the new input text into token IDs
    def encode(self, new_text):
        # Apply the tokenization process to create a list of individual tokens
        text_tokens = re.split(r'([,.?_!"()\']|--|\s)', new_text)
        token_list = [item.strip() for item in text_tokens if item.strip()]

        # Create the list of new token IDs from the list of tokens, using the IDs from the existing vocabulary
        ids = [self.str_to_int[token] for token in token_list]

        return ids
    
    # Create the `decode` method yo convert the token IDs back into text
    def decode(self, ids):
        # Create string of words with spaces, based on their token ID from the existing vocabulary
        text = " ".join([self.int_to_str[id] for id in ids])

        # Remove any spaces before specified punctuation
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)

        return text
```

```python
# Instantiate a new tokenizer object from `SimpleTokenizerV1` on new text
# NOTE: The sample text MUST contain all words from the vocaubulary
tokenizer = SimpleTokenizerV1(vocab)
sample_text = """It's the last painted, you know,
Mrs. Gisburn said with a pardonable pride."""
ids = tokenizer.encode(new_text=sample_text)
print(f"Token IDs:\n{ids}")
```

```python
# Convert the token IDs back into words
print(tokenizer.decode(ids))
```

## Handling "unknown" words and special context tokens

```python
# Rebuild the vocabulary to handle the `<|unk|>` and `<|endoftext|>` tokens
preprocessed = re.split(r'([,.?!;:"()\']|--|\s)', raw_text)
preprocessed = [item.strip() for item in preprocessed if item.strip()]
all_tokens = sorted(list(set(preprocessed)))
all_tokens.extend(["<|endoftext|>", "<|unk|>"])
# NOTE: The vocabulary is in the format of {<WORD_TOKEN> :<TOKEN_ID>}
vocab = {token:integer for integer, token in enumerate(all_tokens)}

# Confirm the new tokens
for index, item in enumerate(list(vocab.items())[-10:]):
    print(item)
```

```python
# Extend the tokenizer class to handle unknown words, using the `<|unk|>` token
class SimpleTokenizerV2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {integer:string for string, integer in vocab.items()}
    
    # Create the `encode` method to process the new input text into token IDs
    def encode(self, new_text):
        text_tokens = re.split(r'([,.?_!"()\']|--|\s)', new_text)
        token_list = [item.strip() for item in text_tokens if item.strip()]

        # Replace unknown words with the `<|unk|>` token
        token_list = [item if item in self.str_to_int else "<|unk|>" for item in token_list]

        return [self.str_to_int[token] for token in token_list]
    
    def decode(self, ids):
        text = " ".join([self.int_to_str[id] for id in ids])
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)

        return text
```

```python
# Test on new text, with words not in the vocabulary
text1 = "Hello, do you like tea?"
text2 = "In the sunlit terraces of the palace."
text = " <|endoftext|> ".join((text1, text2))
print(text)
```

```python
tokenizer = SimpleTokenizerV2(vocab)
print(tokenizer.encode(text))
print(tokenizer.decode(tokenizer.encode(text)))
```

```python
import tiktoken
print(tiktoken.__version__)
```

```python
# Define the tokenizer using BPE
tokenizer = tiktoken.get_encoding("gpt2")
unknown_word = "Akwirw ier"

# Print individual token IDs
print(tokenizer.encode(unknown_word))
```

```python
# View the mapping of how BPE breaks down the sub-words
for id in tokenizer.encode(unknown_word):
    print(f"{id} --> {tokenizer.decode([id])}")
```

```python
# Reconstruct the token IDs to check original worrd
token_ids = []
for id in tokenizer.encode(unknown_word):
    token_ids.append(tokenizer.decode([id]))
new_word = "".join(token_ids)
assert new_word == unknown_word
print(new_word)
```
